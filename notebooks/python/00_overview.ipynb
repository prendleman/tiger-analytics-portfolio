# Healthcare Analytics Portfolio — Overview

This repo demonstrates **Lead Data Scientist** skills end-to-end: synthetic healthcare data, ML (classification, time series, clustering, anomaly detection), SQL, and MLOps-style pipelines.

## Notebooks in this folder

| Notebook | What it does |
|----------|---------------|
| **readmission_prediction.ipynb** | 30-day readmission binary classification (Random Forest + GridSearchCV), feature importance |
| **time_series_pmpm_forecasting.ipynb** | PMPM cost series from claims; SARIMA, exponential smoothing, optional Prophet; MAPE/RMSE |
| **clustering_anomaly_detection.ipynb** | KMeans encounter segments; Isolation Forest for outlier encounters |
